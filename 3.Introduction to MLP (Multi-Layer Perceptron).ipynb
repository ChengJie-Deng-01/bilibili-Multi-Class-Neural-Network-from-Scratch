### Hidden Layer

>- 
>- **RELU (Rectified Linear Unit)** $ \rightarrow f(x) = \max(0, x)$


$$
f(x)=\begin{cases}x, & \text{if } x \geq 0 \\ 0, & \text{if } x < 0\end{cases}
$$

```python
dataset1 = np.array([175, 75, 23, 3000])
dataset1
```

## Step one

```python
h11_weight = np.array([0.08, 0.2, 0.5, 0.001])
h11_bias = 0


h12_weight = np.array([0.01, 0.2, 0.5, -0.08])
h12_bias = 0
```

```python
h1_weight = np.array([[0.08, 0.2, 0.5, 0.001],
                     [0.01, 0.2, 0.5, -0.08]])

h1_weight.T


h1_bias = np.array([0, 0])
```

```python
input1 = np.dot(dataset1, h1_weight.T) + h1_bias
input1
```

## Step two

```python
h1_output = np.maximum(0, input1)
h1_output
```

```python
input2 = h1_output
input2
```

## Step three

```python
output_weight = np.array([[0.01, 0.05],
                         [0.03, -0.6],
                         [0.03, -0.9]])

output_bias = np.array([0, 0, 0])

np.dot(input2, output_weight.T) + output_bias
```
