### Hidden Layer

>- 隐藏层的作用是让网络具有“非线性建模能力”，能学到更复杂的规律和特征
>- 隐藏层后面接非线性激活函数，比如**RELU (Rectified Linear Unit)** $ \rightarrow f(x) = \max(0, x)$


$$
f(x)=\begin{cases}x, & \text{if } x \geq 0 \\ 0, & \text{if } x < 0\end{cases}
$$

```python
dataset1 = np.array([175, 75, 23, 3000])
dataset1
```

## Step one

```python
h11_weight = np.array([0.08, 0.2, 0.5, 0.001])
h11_bias = 0


h12_weight = np.array([0.01, 0.2, 0.5, -0.08])
h12_bias = 0
```

```python
h1_weight = np.array([[0.08, 0.2, 0.5, 0.001],
                     [0.01, 0.2, 0.5, -0.08]])

h1_weight.T


h1_bias = np.array([0, 0])
```

```python
input1 = np.dot(dataset1, h1_weight.T) + h1_bias
input1
```

## Step two

```python
h1_output = np.maximum(0, input1)
h1_output
```

```python
input2 = h1_output
input2
```

## Step three

```python
output_weight = np.array([[0.01, 0.05],
                         [0.03, -0.6],
                         [0.03, -0.9]])

output_bias = np.array([0, 0, 0])

np.dot(input2, output_weight.T) + output_bias
```

### Summary

| 特征   | Perceptron     | MLP（多层感知机）              |
| ---- | -------------- | ----------------------- |
| 层数   | 1（没有隐藏层）       | 至少有 1 个隐藏层              |
| 输出   | 通常只有 1 个输出     | 可以有多个输出（多类、多标签任务）       |
| 激活函数 | 通常是 sign（符号函数） | 使用非线性激活（如 ReLU、sigmoid） |
| 表达能力 | 只能处理线性可分问题     | 可以学习非线性关系，适合复杂任务        |
