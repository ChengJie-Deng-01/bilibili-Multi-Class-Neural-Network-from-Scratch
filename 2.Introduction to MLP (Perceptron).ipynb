## 1: Perceptron (感知机)

>- 是最早的人工神经网络模型之一，最初的形式是一个只有一层神经元的二分类模型
>- 没有隐藏层，只有输入层和输出层（输出层通常只有一个神经元）

```python
Image(filename='/Users/chengjiedeng/Desktop/bilibil-MLP/2.MLP/Perceptron.png', width=600)
```

假设我们的输入特征为：

$$
x = [x_1 = 172, x_2 = 75, x_3 = 23, x_4 = 3000] \rightarrow [身高, 体重, 年龄, 工资]
$$

每个输入都有对应的权重参数：

$$
w = [w_1 = 0.08, w_2 = 0.2, w_3 = 0.5, w_4 = 0.001]
$$

还有一个偏置项：

$$
b = 0
$$

```python
dataset1 = np.array([175, 75, 23, 3000])
dataset1
```

```python
output_weight = np.array([0.08, 0.2, 0.5, 0.001])
```

```python
np.dot(dataset1, output_weight)
```

### 1️⃣ 线性组合（加权求和 + 偏置）

将输入与权重一一相乘后相加，再加上偏置：

$$
z = w_1x_1 + w_2x_2 + w_3x_3 + w_4x_4 + b = \mathbf{w}^\top \mathbf{x} + b
$$

这是感知机的“感知”过程，用于衡量输入的重要性。



```python
175*0.08 + 75*0.2 + 23*0.5 + 3000*0.001
```

```python
input1 = 175*0.08 + 75*0.2 + 23*0.5 + 3000*0.001 + 0
input1
```

### 2️⃣ 激活函数（Step Function）

感知机使用的是**符号函数**（sign function，也叫 step function）作为激活函数：

$$
\hat{y} =
\begin{cases}
1 & \text{if } z \ge 0 \\
-1 & \text{if } z < 0
\end{cases}
$$

这个函数是一个**硬分类器**，只输出 0 或 1，用于二分类。



```python
if input1 >= 0:
    
    output = 1
    
else:
    
    output = -1
```

```python
output
```




### 3️⃣ 输出结果


$$
\hat{y} = \text{sign}(\mathbf{w}^\top \mathbf{x} + b)
$$
