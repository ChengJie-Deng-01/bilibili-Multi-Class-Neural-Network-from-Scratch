## 1: Forward Propagation


78428x28  

$$
\mathbf{x} = [x_1, x_2, \dots, x_{784}]
$$


```python
import matplotlib.pyplot as plt
import numpy as np

plt.imshow(np.array([[0, 255],
                     [255, 0]]), cmap='gray', vmin=0, vmax=255)
```

## 2: Hidden Layer

10****

$$
z^{(1)}_j = \sum_{i=1}^{784} w^{(1)}_{ji} \cdot x_i + b^{(1)}_j
$$



## 3: Activation Function

 ReLU

$$
a^{(1)}_j = \text{ReLU}(z^{(1)}_j)
$$

## 4: Output Layer

100~9

$$
z^{(2)}_k = \sum_{j=1}^{10} w^{(2)}_{kj} \cdot a^{(1)}_j + b^{(2)}_k
$$

## 5: Softmax 



$$
\hat{y}_k = \frac{e^{z^{(2)}_k}}{\sum_{l=1}^{10} e^{z^{(2)}_l}}
$$

 $\hat{y}_k$  $k$  $\hat{y}$ 1


## 6: 



$$
\text{predicted class} = \arg\max_k \hat{y}_k
$$
